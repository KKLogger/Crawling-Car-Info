```python
tr_list = soup.find_all('tr')
for tr in tr_list:
    if tr.find('th',{'scope':'row'}) is not None:
        key = tr.find('th',{'scope':'row'}).text
        td = tr.find('td')
        if td.find('div',{'class','option-ch'}) is None:
            value = td.text.strip()
        else:
            if td.find('div',{'class','option-ch'}).get('value') is not None :
                if td.find('div',{'class','option-ch'})['value'] == ' ' or td.find('div',{'class','option-ch'})['value'] == '' :
                    value ="None"
                else:
                    value = td.find('div',{'class','option-ch'})['value']
            else:        
                value  = 'None'
        print( key,"/", value)
```
